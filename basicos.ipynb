## Ignorando as chaves e utilizando a posição como referência e pode ser usado com fateamento

## Mudando os índices e vinculando ao valor 

```python
indices = [
    "Sandro", "Paula", "Priscila"
]
```

```python
serie = pd.Series(numeros, index=indices)
```

```python
serie
```

```python
print(serie["Sandro"])
```

o **iloc** navega pelas linhas (número da linha) que não é a mesma coisa índices

```python
print(serie.iloc["Sandro"])
```

Outra possibilidade de usar o índice, também pode ser utilizado o loc. Na prática terá o mesmo efeito da busca como uma lista.

```python
print(serie.loc["Paula"])
```
